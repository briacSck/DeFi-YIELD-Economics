```python
"""
Initial Yield Landscape Analysis
Using comprehensive DeFi data with computed metrics
"""

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (16, 10)
plt.rcParams['font.size'] = 11

print("✓ Imports loaded")
```

```python
# Load data
df = pd.read_csv('../data/raw/defi_yields_latest.csv')

print(f"✓ Loaded {len(df):,} pools")
print(f"  Chains: {df['chain'].nunique()}")
print(f"  Protocols: {df['protocol'].nunique()}")
print(f"  Date: {df['date'].iloc[0]}")

# Quick data check
df.head()
```

```python
# Create comprehensive visualization
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('DeFi Stablecoin Yield Landscape Analysis - Feb 2026', 
             fontsize=16, fontweight='bold')

# ============================================================================
# CHART 1: APY Distribution by Protocol (Top 15)
# ============================================================================

top_protocols = df.groupby('protocol')['apy_total'].mean().nlargest(15).index
df_top = df[df['protocol'].isin(top_protocols)]

ax1 = axes[0, 0]
df_top.boxplot(column='apy_total', by='protocol', ax=ax1, rot=90)
ax1.set_title('APY Distribution: Top 15 Protocols')
ax1.set_ylabel('APY (%)')
ax1.set_xlabel('')
plt.sca(ax1)
plt.xticks(rotation=45, ha='right')

# ============================================================================
# CHART 2: Risk-Adjusted Returns
# ============================================================================

ax2 = axes[0, 1]
df_sharpe = df[df['sharpe_proxy'].notna() & np.isfinite(df['sharpe_proxy'])]
top_sharpe = df_sharpe.nlargest(20, 'sharpe_proxy')

scatter = ax2.scatter(top_sharpe['sharpe_proxy'], 
                     top_sharpe['apy_total'],
                     s=top_sharpe['tvl_millions']*5,
                     c=top_sharpe['apy_total'],
                     cmap='viridis',
                     alpha=0.6)

for idx, row in top_sharpe.head(5).iterrows():
    ax2.annotate(f"{row['protocol']}\n{row['chain']}", 
                (row['sharpe_proxy'], row['apy_total']),
                fontsize=8, ha='center')

ax2.set_xlabel('Risk-Adjusted Return (Sharpe Proxy)')
ax2.set_ylabel('APY (%)')
ax2.set_title('Protocol Efficiency Frontier\n(bubble size = TVL)')
plt.colorbar(scatter, ax=ax2, label='APY (%)')

# ============================================================================
# CHART 3: Chain Comparison
# ============================================================================

ax3 = axes[0, 2]
chain_type_stats = df.groupby('chain_type')['apy_total'].agg(['mean', 'count'])
chain_type_stats = chain_type_stats.sort_values('mean', ascending=False)

bars = ax3.barh(chain_type_stats.index, chain_type_stats['mean'])
ax3.set_xlabel('Average APY (%)')
ax3.set_title('Average APY by Chain Type')

for i, (idx, row) in enumerate(chain_type_stats.iterrows()):
    ax3.text(row['mean'] + 0.2, i, f"n={int(row['count'])}", 
            va='center', fontsize=9)

# ============================================================================
# CHART 4: Protocol Tier Performance
# ============================================================================

ax4 = axes[1, 0]
tier_stats = df.groupby('protocol_tier')['apy_total'].agg(['mean', 'median', 'count'])
tier_stats = tier_stats.sort_values('mean', ascending=True)

x = np.arange(len(tier_stats))
width = 0.35

bars1 = ax4.barh(x - width/2, tier_stats['mean'], width, label='Mean', alpha=0.8)
bars2 = ax4.barh(x + width/2, tier_stats['median'], width, label='Median', alpha=0.8)

ax4.set_yticks(x)
ax4.set_yticklabels(tier_stats.index)
ax4.set_xlabel('APY (%)')
ax4.set_title('Performance by Protocol Maturity')
ax4.legend()

# ============================================================================
# CHART 5: Stablecoin Type Comparison
# ============================================================================

ax5 = axes[1, 1]
stable_stats = df.groupby('stable_type').agg({
    'apy_total': 'mean',
    'tvl_usd': 'sum'
}).sort_values('apy_total', ascending=False)

bars = ax5.bar(stable_stats.index, stable_stats['apy_total'])
ax5.set_ylabel('Average APY (%)')
ax5.set_title('APY by Stablecoin Type')
ax5.tick_params(axis='x', rotation=45)

ax5_2 = ax5.twinx()
ax5_2.plot(stable_stats.index, stable_stats['tvl_usd']/1e9, 
          'ro-', linewidth=2, markersize=8, label='Total TVL')
ax5_2.set_ylabel('Total TVL ($B)', color='r')
ax5_2.tick_params(axis='y', labelcolor='r')

# ============================================================================
# CHART 6: TVL vs APY Scatter
# ============================================================================

ax6 = axes[1, 2]

top_tvl_protocols = df.groupby('protocol')['tvl_usd'].sum().nlargest(30).index
df_tvl = df[df['protocol'].isin(top_tvl_protocols)]

for protocol in df_tvl['protocol'].unique()[:10]:
    proto_data = df_tvl[df_tvl['protocol'] == protocol]
    ax6.scatter(proto_data['tvl_millions'], 
               proto_data['apy_total'],
               label=protocol,
               alpha=0.6,
               s=50)

ax6.set_xlabel('TVL ($ Millions)')
ax6.set_ylabel('APY (%)')
ax6.set_title('TVL vs APY: Where is Smart Money?')
ax6.set_xscale('log')
ax6.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)

plt.tight_layout()
plt.savefig('../results/figures/yield_landscape_feb2026.png', dpi=300, bbox_inches='tight')
print("✓ Saved: results/figures/yield_landscape_feb2026.png")

plt.show()
```

```python
# Generate insights
print("\n" + "="*70)
print("💡 KEY RESEARCH INSIGHTS")
print("="*70)

# Finding 1: Best risk-adjusted protocols
print("\n1️⃣ BEST RISK-ADJUSTED PROTOCOLS (Top 5 Sharpe):")
top5_sharpe = df[df['sharpe_proxy'].notna()].nlargest(5, 'sharpe_proxy')
for idx, row in top5_sharpe.iterrows():
    print(f"   • {row['protocol']:20} on {row['chain']:15}")
    print(f"     {row['apy_total']:5.2f}% APY | Sharpe: {row['sharpe_proxy']:5.2f}")

# Finding 2: Chain performance
print("\n2️⃣ CHAIN TYPE PERFORMANCE:")
chain_means = df.groupby('chain_type')['apy_total'].mean().sort_values(ascending=False)
for chain_type, apy in chain_means.items():
    print(f"   • {chain_type:20}: {apy:5.2f}% avg")

# Finding 3: Tier analysis
print("\n3️⃣ PROTOCOL MATURITY TRADE-OFF:")
tier_analysis = df.groupby('protocol_tier').agg({
    'apy_total': ['mean', 'std'],
    'sharpe_proxy': 'mean'
}).round(2)
tier_analysis.columns = ['apy_mean', 'apy_std', 'sharpe_mean']
for tier, row in tier_analysis.iterrows():
    print(f"   • {tier:25}: {row['apy_mean']:5.2f}% ± {row['apy_std']:4.2f}% "
          f"| Sharpe: {row['sharpe_mean']:5.2f}")

# Finding 4: Validate 8-15% claim
print("\n4️⃣ APY CLAIM VALIDATION:")
avg_apy = df['apy_total'].mean()
median_apy = df['apy_total'].median()
print(f"   Market average: {avg_apy:.2f}%")
print(f"   Market median: {median_apy:.2f}%")
if 8 <= avg_apy <= 15:
    print(f"   ✓ VALIDATED: Within 8-15% target range")
else:
    print(f"   ⚠️  Outside target: Adjust product claims")

# Finding 5: Optimal allocation
print("\n5️⃣ OPTIMAL ALLOCATION RECOMMENDATION:")
liquid_pools = df[(df['tvl_millions'] > 10) & df['sharpe_proxy'].notna()]
top_allocation = liquid_pools.nlargest(3, 'sharpe_proxy')
print("   Suggested portfolio (by risk-adjusted return + liquidity):")
for i, row in enumerate(top_allocation.iterrows(), 1):
    idx, data = row
    print(f"   {i}. {data['protocol']:18} ({data['asset']:6}) on {data['chain']}")
    print(f"      {data['apy_total']:5.2f}% APY | ${data['tvl_millions']:.1f}M TVL")

print("\n" + "="*70)
```
